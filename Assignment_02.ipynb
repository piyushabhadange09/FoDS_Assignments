```python
print("1ST PLOT: PIE CHART")
print("Pie Chart of Iris dataset:")
data = df['type'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(data, labels=data.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Flower Types')
plt.show()
```

```python
print("2) SCATTER PLOT")
print("Scatter plot between Sepal Length and Sepal Width:");
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='x1', y='x2', hue='type')
plt.title('Scatter Plot of Sepal Length vs. Sepal Width')
plt.xlabel('Sepal Length (x1)')
plt.ylabel('Sepal Width (x2)')
plt.legend(title='Flower Type')
plt.show()
```

```python
print("Scatter plot between Petal Length and Petal Width:");
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='x3', y='x4', hue='type')
plt.title('Scatter Plot of Petal Length vs. Petal Width')
plt.xlabel('Petal Length (x3)')
plt.ylabel('Petal Width (x4)')
plt.legend(title='Flower Type')
plt.show()
```

Setosa has smallest Petal Length and Petal Width,Versicolor has average petal length and petal width and virginica has largest petal length and petal width.

```python
print("3) BAR PLOT")
print("Bar Plot of Iris Dataset:")
df.plot.bar()
plt.title("Bar Plot")
```

A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally.

```python
print("Bar Graph of Iris Dataset when stacked is allowed:")
# We can also create stacked bar chart. Stacked bar plots are an excellent way to visualize data that is divided into different categories.
df.plot.bar(stacked=True)
plt.title("Stacked Bar Plot")
```

```python
print("4)Area Plot")
print("Area Plot of Iris Dataset")
df.plot.area(alpha=0.7)
plt.title("Area Plot")
```

An area chart or area graph displays graphically quantitative data. It is based on the line chart. The area between axis and line are commonly emphasized with colors, textures and hatchings.

```python
print("5)Density Plot-(KDE -Kernel Density Estimation)")
print("density plot of sepal length(x1)")
#density plot of x1 using kde
df['x1'].plot.kde()
plt.title("density plot of sepal length")
```

```python
print("density plot of sepal width(x2)")
df['x2'].plot.kde()
plt.title("density plot of sepal width")
```

```python
print("density plot of petal length(x3)")
df['x3'].plot.kde()
plt.title("density plot of petal length")
```

```python
df['x4'].plot.kde()
plt.title("density plot of petal width")
```

```python
print("Box plot of iris dataset:")
df.plot.box() 
plt.title("Box Plot")
```

A boxplot is a standardized way of displaying the distribution of data based on a five-number summary (“minimum”, first quartile (Q1), median, third quartile (Q3), and “maximum”).

```python
print("Hexagonal plot")
df.plot.hexbin(x='x1', y='x2', gridsize=20, cmap='Blues')
plt.title('Hexagonal Bin Plot of Sepal Length vs Sepal Width')
plt.show()
```

Why to use Hexa Bin Graph:
1.Each hexagon represents a region where points are grouped. The color intensity of the hexagon shows how many data points fall within that area.
2.Darker hexagons indicate areas where more observations have similar values for sepal length and sepal width.
3.Lighter hexagons indicate areas with fewer observations.

```python
df.plot.hexbin(x='x3', y='x4', gridsize=20, cmap='Oranges')
plt.title('Hexagonal Bin Plot of Petal Length vs Petal Width')
plt.show()
```
