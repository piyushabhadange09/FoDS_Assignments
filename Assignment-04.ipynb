```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Parameters for the normal distribution
mean=100
std_deviation=4
dataset_size = 100000
```

```python
#Generating random data using normal distribution
data = np.random.normal(loc=mean, scale=std_deviation, size=dataset_size)
data
```

```python
#visulization of normal distribution using seaborn library
plt.figure(figsize=(10, 6))
sns.histplot(data, bins=50, kde=True, color='skyblue',edgecolor='black')
plt.title('Normal Distribution (Mean = 100, Std Dev = 4)', fontsize=20)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.grid(True)
plt.show()
```
